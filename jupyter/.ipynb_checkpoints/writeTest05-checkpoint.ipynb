```python
# 병합(merge) : 최소 2개 이상의 데이터프레임을 합치는 작업
import pandas as pd

# 예시 데이터
df1 = pd.DataFrame({
    "지역":["서울","부산","대구"],
    # "연도":[2024,2025,2024],
    "총인구":[1000000,300000,200000]
})
df2 = pd.DataFrame({
    "지역":["서울","부산","인천"],
    # "연도":[2024,2025,2024],
    "병원수":[150,80,60]
})
# 데이터프레임의 필드명 확인하는 메서드
print(df1.columns)
print(df2.columns)

# pd.merge(left_DataFrameName, right_DataFrameName, how="타입", on="기준컬럼")
# how 방식 : inner(교집합), left(왼쪽기준), right(오른쪽 기준), outer(합집합)

# 01. 교집합
# merged = pd.merge(df1,df2,how='inner',on='지역') # 서울, 부산만 출력

# 02. 합집합
# merged = pd.merge(df1,df2,how='outer',on='지역') # 전부 출력하되, NaN출력

# 03. left(왼쪽기준)
# merged = pd.merge(df1,df2,how='left',on='지역') # 서울,부산,대구만 출력(왼쪽만)

# 03. right(오른쪽기준)
merged = pd.merge(df1,df2,how='right',on='지역') # 서울,부산,인천만 출력(오른쪽만)

# 컬럼(필드)명이 다른 경우
# merged = pd.merge(df1,df2,how='inner',left_on='지역',right_on='지역명')

# 여러개의 컬럼(필드)명으로 병합하는 경우
# merged = pd.merge(df1,df2,how='inner',on=['지역','연도'])
print(merged)
```

```python
# 결측치란 : 데이터 병합 과정에서 데이터가 비어있는 상태를 의미
# 즉, 값이 없어서 NaN(Not a Number)로 나타나는 경우

# 예시 데이터
df1 = pd.DataFrame({
    "지역":["서울","부산","대구"],
    "총인구":[1000000,300000,200000]
})
df2 = pd.DataFrame({
    "지역":["서울","부산","인천"],
    "병원수":[150,80,60]
})

# 병합
merged02 = pd.merge(df1,df2,how='outer',on='지역')
# print(merged02)

# 결측 개수 확인
# isna() : 결측치가 존재
# print(merged02.isna().sum())

# 결측치 처리 - 3가지 방식
# 1) 특정 값으로 채우기 : fillna(0), fillna(평균값) 등
# 총 인구와 병원수 결측치 0으로 채우기
# merged02['총인구'] = merged02['총인구'].fillna(0)
# merged02['병원수'] = merged02['병원수'].fillna(0)

# 2) 행 삭제하기 : dropna -> 결측치가 존재하는 행을 통째로 삭제
merged02 = merged02.dropna()

print(merged02)
```
