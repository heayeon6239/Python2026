```python
# 01. 반드시 pandas는 import하여 사용
import pandas as pd
# matplotlib 매트플롯립도 반드시 import하여 사용
import matplotlib.pyplot as plt

# 02. 데이터 생성하기
df1 = pd.DataFrame({
    "지역": ["서울", "부산", "대구"],
    "총인구": [1000000,300000, 200000]
})

df2 = pd.DataFrame({
    "지역": ["서울", "부산", "인천"],
    "병원수": [150,80, 60]
})

# 03. 데이터 전처리를 위해 merge(병합)
# outer로 병합하면 df1과 df2의 합집합
merged = pd.merge(df1, df2, on="지역", how="outer")
merged
# 04. 결측치 처리
merged["총인구"] = merged["총인구"].fillna(merged["총인구"].mean())
merged["병원수"] = merged["병원수"].fillna(merged["병원수"].mean())

# -----------------------------
# 세로막대 그래프

# 01. 한글 깨짐 방지 설정★
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 차트 배경 + 색상 자동으로 그리드에 빨간색으로 지정
# plt.style.use('ggplot')

# 02. 그래프 크기 설정
plt.figure(figsize=(8,5))

# 03. 세로그래프(x축: 지역, y축: 병원수)
# 세로막대 넓이 기본값 : width=0.8
bars = plt.bar(merged["지역"],merged["병원수"],width=0.4, color="blue")

# 04. 차트제목, 가로축, 세로축 제목 지정
plt.title("지역별 병원수")
plt.xlabel("지역")
plt.ylabel("병원수")

for bar in bars:
    height = bar.get_height() # 막대 높이 추출하는 함수 예) 대구 : 96
    # bar.get_x() : 막대그래프 넓이의 시작점 추출
    # bar.get_width() : 막대그래프 넓이전체 추출
    plt.text(bar.get_x() + bar.get_width()/2, height, 
             f'{int(height)}', ha='center', va='bottom')

# 05. 저장
plt.savefig('hospital_chart.png',dpi=300)

# 06. 그래프 출력
# plt.show()
```

```python
# 01. 반드시 pandas는 import하여 사용
import pandas as pd
# matplotlib 매트플롯립도 반드시 import하여 사용
import matplotlib.pyplot as plt
# seaborn : matplotlib 기반위에서 차트 이쁘게 꾸미는 도구
import seaborn as sns

# 02. 데이터 생성하기
df1 = pd.DataFrame({
    "지역": ["서울", "부산", "대구"],
    "총인구": [1000000,300000, 200000]
})

df2 = pd.DataFrame({
    "지역": ["서울", "부산", "인천"],
    "병원수": [150,80, 60]
})

# 03. 데이터 전처리를 위해 merge(병합)
# outer로 병합하면 df1과 df2의 합집합
merged = pd.merge(df1, df2, on="지역", how="outer")
merged
# 04. 결측치 처리
merged["총인구"] = merged["총인구"].fillna(merged["총인구"].mean())
merged["병원수"] = merged["병원수"].fillna(merged["병원수"].mean())

# -----------------------------
# seaborn 스타일 지정(한글 깨짐 방지 설정 위에 존재해야 한글이 깨지지 않음★)
sns.set_style('whitegrid') # 배경 스타일 설정, dark : 그리드 없이 어두운 배경 지정
sns.set_palette('pastel') # 색상 스타일 지정, deep : 진하게

# 01. 한글 깨짐 방지 설정★
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 차트 배경 + 색상 자동으로 그리드에 빨간색으로 지정
# plt.style.use('ggplot')

# 02. 그래프 크기 설정
plt.figure(figsize=(8,5))

# 03. 세로그래프(x축: 지역, y축: 병원수)
# 세로막대 넓이 기본값 : width=0.8
# bars = plt.bar(merged["지역"],merged["병원수"],width=0.4, color="blue")
ax = sns.barplot(
    x='지역',
    y='병원수',
    data=merged, 
    width=0.4, 
    # color='skyblue'
    palette=['red','green','blue','yellow']
)

# 04. 차트제목, 가로축, 세로축 제목 지정
plt.title("지역별 병원수",fontsize=18, fontweight='bold', color='darkblue', pad=20) # 차트 제목 여백
plt.xlabel("지역",fontsize=14, fontweight='bold', color='purple')
plt.ylabel("병원수",fontsize=14, fontweight='bold', color='orange')

# x축 눈금의 값 각종 글꼴 지정하기
plt.xticks(fontsize=12, fontweight='bold', color='brown')
# y축 눈금의 값 각종 글꼴 지정하기
plt.yticks(fontsize=12, fontweight='bold', color='black')

# 세로막대 각각의 데이터 레이블 값 출력 방법
# for bar in bars:
#     height = bar.get_height() # 막대 높이 추출하는 함수 예) 대구 : 96
#     # bar.get_x() : 막대그래프 넓이의 시작점 추출
#     # bar.get_width() : 막대그래프 넓이전체 추출
#     plt.text(bar.get_x() + bar.get_width()/2, height, 
#              f'{int(height)}', ha='center', va='bottom')
for i in range(len(merged['병원수'])):
    v = merged['병원수'][i] # i번째 병원수 값 출력
    print(f'i번째 병원수:{v}')
    ax.text(
        i, # x의 위치(막대위치: 0,1,2...)
        v, # y의 위치 (막대높이)
        int(v), # 정수로 표시할 값
        ha='center',
        va='bottom',
        fontsize=12,
        fontweight='bold',
        color='red'
    )

# 차트 그리드 선모양 지정하는 방법
# plt.grid(linestyle='--') # 점선
plt.grid(linestyle=':') # 촘촘한 선
# plt.grid(linestyle='-.') # 점+선 혼합

# 그래프 요소들이 겹치지 않게 자동으로 간격을 조정해주는 함수
plt.tight_layout()

# 06. 그래프 출력
# plt.show()
```
