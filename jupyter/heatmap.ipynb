```python
# 01. 반드시 pandas는 import하여 사용
import pandas as pd
# matplotlib 매트플롯립도 반드시 import하여 사용
import matplotlib.pyplot as plt
import seaborn as sns

# 01. 한글 깨짐 방지 설정★
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

data = pd.DataFrame({
    "수학":[80,90,70],
    "영어":[85,60,75],
    "과학":[90,88,95],
},index=['김학생','이학생','박학생'])

# heatmap() : 데이터 값을 색으로 표현하는 그래프
# 숫자가 클수록 -> 연한색
# 숫자가 작을수록 -> 진한색
sns.heatmap(data)
plt.title('학생 성적 히트맵')
plt.show()
```
