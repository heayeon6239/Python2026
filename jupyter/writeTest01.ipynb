## utf-8-sig는 윈도우 환경에서 텍스트 파일 저장시 앞 부분에 기호가 같이 저장되는 부분을 해결해주는 인코딩

```python
with open('greenData/text_data.txt','w',encoding='utf-8-sig') as f:
    f.write('데이터 분석을 위한 첫 걸음\n')
    f.write('파이썬 기본 파일 입출력 실습입니다.\n')
    f.write('1. 기초 문법 익히기\n')
    f.write('2. 라이브러리 활용하기\n')
    f.write('3. 시각화 진행하기\n')
```

문제) score_data.csv 파일의 데이터 중 국어점수가 85점 이상인 데이터만 추출하시오.
## DictReader => csv 파일을 읽기 좋게 변환하는데 첫 줄을 제목(=필드명)으로 인식하도록 변경해줌

```python
import csv
# 01. 파일 읽기
with open('greenData/score_data.csv','r',encoding='utf-8-sig') as f:
    # 02. CSV의 첫줄을 필드명으로 인식하도록 DictReader 이용
    reader = csv.DictReader(f)
    
    print('국어 점수가 85점 이상인 데이터: ')

    # 03. 한줄씩 확인하면서 if조건 검사
    for r in reader:
        # 파일에서 받아온 모든 숫자는 문자열이므로, 반드시 int()로 숫자형 변환★
        if int(r['국어']) >= 85:
            print(r)
```
