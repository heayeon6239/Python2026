```python
# 공공데이터포털 사이트에서 다운로드 받은 파일 읽어오는 방법
df = pd.read_csv('greenData/pm10.csv',encoding='cp949')
df
```

```python
# 구분 컬럼의 데이터 중'평균'인 행을 제외하여 실제 자치구 데이터만 남깁니다. (데이터전처리)
data = df[df['구분'] != '평균']
# 미세먼지(PM10) 농도가 25 이상이면서 동시에 초미세먼지(PM25) 농도가 15 이상인 '주의 요망' 지역 찾기
# 다중 조건 줄 때 and : &, or : **, not : ~
data02 = (data['미세먼지(PM10)'] >= 25) & (data['초미세먼지(PM25)'] >= 15)
data03 = data[data02]
result = data03[['구분','미세먼지(PM10)','초미세먼지(PM25)']]
print(result)
```

```python
# from datetime import datetime (import시 그냥 datetime만 작성)

# 공공데이터포털 사이트에서 다운로드 받은 파일 읽어오는 방법
df = pd.read_csv('greenData/accident.csv',encoding='cp949')
df
```

```python
# pandas에서 날짜를 추출하는 함수 종류
# CSV에 입력된 날짜는 문자열 데이터 타입
# ★ 날짜형식으로 형변환하는 함수 : datetime()

# 사고일자 컬럼을 datetime 형식으로 변환하고, 이를 활용하여'사고월'이라는 새로운 파생변수(컬럼)를 생성
# 사고일자가 날짜형식으로 변경됨
df['사고일자'] = pd.to_datetime(df['사고일자'])
df
```

```python
# 문제) 사고일자 컬럼을 datetime 형식으로 변환하고, 이를 활용하여'사고월'이라는 새로운 파생변수(컬럼)를 생성

# (1)사고일자의 월을 출력 : dt.month(월), dt.year(년), dt.day(일), dt.hour(시), dt.minute(분), dt.second(초)
# (2) dt.dayofweek : 요일을 숫자로 반환(0:월요일 ~ 6:일요일)
# (3) dt.day_name() : 요일 이름이 문자열('Monday')
# (4) dt.quarter : 분기(1분기 ~ 4분기)
df['사고월'] = df['사고일자'].dt.month
df
```

```python
# 문제) 월별 사고 건수를 집계하여 어느 월에 사고가 가장 많이 발생했는지 확인하세요.
# groupby().count() => 데이터가 비어있지 않은 값의 개수를 셈
# groupby().size() => 그룹별 전체 행의 개수를 셈 (비어있는 값의 개수도 셈)
monthCount = df.groupby('사고월').size()
# idxmax() => 가장 큰 값이 들어있는 위치(인덱스 이름)에 접근하여 가져옴 -> 가장 큰 값의 인덱스를 가져옴★
maxMonth = monthCount.idxmax()
print(monthCount)
print(f"가장 많이 발생한 월: {maxMonth}월")
```

```python
# 문제) 사망자수와 부상자수를 더하여 '사상자수'라는 파생변수를 생성하세요
df['사상자수'] = df['사망'] + df['부상']
df
```

```python
result = df.groupby('노선명')['사상자수'].sum().sort_values(ascending=False).head(5)
print(result)
```
